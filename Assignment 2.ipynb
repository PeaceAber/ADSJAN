```python
###number one
import pandas as pd
import plotly.graph_objects as go
atr = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
atr.head()

trace0 = go.Box(
    x=atr.JobRole,
    y=atr.DistanceFromHome,
    name = 'Job role',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)

data = [trace0]

layout = go.Layout(title='Distribution of distance by job role',
                   xaxis=dict(title='Job role'),
                   yaxis=dict(title='Distance')
                  )
fig = go.Figure(data=data, layout = layout)


#fig = go.Figure(dict(data = data, layout = layout))

fig.show()
```
